# Seasonality Data Cleaning

Data cleaning for Airbnb seasonality file.

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.cm as cmx
import matplotlib.colors as colors
import numpy as np
%matplotlib inline
```

## Overall Goal:
The goal of this part of the cleaning is to get the dataset in a usable format. We ran into a lot of problems because our csv contained commas in the pricing column, thus creating a problem with the number of columns for select entires(where the price is 1,000+). A lot of time had to be spent to recreating the csv in a usable format. Here we do this by using various strip and string concatenate functions and then output the correct file. Lastly, we see ony a small subset of date don't have prices. It doesn't make much sense to predict those seasonality prices and then run a seasonality analysis. Given it is a small number of the overall entries, we drop the null values as so it doesn't cause errors with our seasonality analysis.

```python
data = pd.read_csv('../datasets/raw_datasets/calendar.csv')
data['Unnamed: 4'].fillna(value='',inplace=True)
price_strip = data['price'].str.strip()
data['new price']= price_strip.map(str)+data['Unnamed: 4'].map(str)
data[data['new price']=='nan']=np.NaN
data_fill = data.dropna()
data_fill = data_fill.drop('price', 1)
data_fill.drop("Unnamed: 4",axis=1,inplace=True)
data_fill.head(5)
```

We see that everything looks great! We export it to the file calendar_clean.csv

```python
data_fill.to_csv("../datasets/calendar_clean.csv",index=False)
```
